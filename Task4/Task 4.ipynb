```python
#Import required libraries
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
```

```python
# Import necessary libraries for graph
from io import StringIO 
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
```

```python
# Load the iris dataset
iris = pd.read_csv(r'C:\Users\Vinu\Downloads\Iris.csv')
```

```python
#for coding of target variable
cleanup_nums = {"Species":     {"Iris-setosa": 1, "Iris-versicolor": 2,"Iris-virginica": 3}}
cleanup_nums
iris.replace(cleanup_nums, inplace=True)
iris.head()
```

```python
#To create Dataframe
iris_df = pd.DataFrame(data=iris)
iris_df
```

```python
iris_df.drop('Id',axis=1)
```

```python
#To select target and features variables
y=iris_df['Species']
x=iris_df.drop(['Species','Id'],axis=1)
```

```python
#Training and testing
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)
```

```python
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(x,y)
```

```python
# Visualize the graph
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
